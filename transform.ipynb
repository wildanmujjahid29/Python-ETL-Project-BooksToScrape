## Transform Data Hasil Scrape

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/books_kotor.csv')
df
```

```python
df.info()
```

```python
df = df[df['Title'].str.strip().str.lower() != 'unknown product']
df
```

```python
def convert_price(price_str):
    try:
        price = float(price_str.replace('£', '').replace(',', ''))
        return int(price * 18402.00)
    except:
        return np.nan
df['Price'] = df['Price'].apply(convert_price)
```

```python
df['Price'].head(5)
```

```python
df['Rating'].value_counts()
```

```python
# Mapping Rating
def map_rating(rating_str):
    mapping = {
        'One': 1,
        'Two': 2,
        'Three': 3,
        'Four': 4,
        'Five': 5
    }
    return mapping.get(rating_str, np.nan)

df['Rating'] = df['Rating'].apply(map_rating)
```
